```python
def transcribe_DNA(sequences, tipe):
  new_sequence = ''
  for sequence in sequences:
    if tipe == 'Intermediate':
      new_sequence += 'C' if sequence == 'G' else 'G' if sequence == 'C' else 'T' if sequence == 'A' else 'A'

    if tipe == 'RNA messanger':
      new_sequence += 'U' if sequence == 'T' else sequence
  return new_sequence

def split_codons(sequences):
  mrna_seq = sequences
  spliting = len(sequences) % 3
  if spliting > 0:
    mrna_seq = sequences[:-spliting]

  return [mrna_seq[i:i+3] for i in range(0, len(mrna_seq), 3)]
```
