```python
def transcribe_DNA(sequences, tipe):
  new_sequence = ''
  for sequence in sequences:
    if tipe == 'Intermediate':
      new_sequence += 'C' if sequence == 'G' else 'G' if sequence == 'C' else 'T' if sequence == 'A' else 'A'

    if tipe == 'RNA messanger':
      new_sequence += 'U' if sequence == 'T' else sequence
  return new_sequence

def split_codons(sequences):
  mrna_seq = sequences
  spliting = len(sequences) % 3
  if spliting > 0:
    mrna_seq = sequences[:-spliting]

  return [mrna_seq[i:i+3] for i in range(0, len(mrna_seq), 3)]
```

```python
def blosum62_score(aa1, aa2):
  blosum62 = blossum62_matrix()
  # the pinalty is -1
  return blosum62.get((aa1, aa2), blosum62.get((aa2, aa1), -1))

def normalization_length(seqf, seqs):
  mySeq = []
  if len(seqf) < len(seqs):
    seqf = [seqf[i] if i < len(seqf) else ' ' for i in range(len(seqs))]

  if len(seqf) > len(seqs):
    seqs = [seqs[i] if i < len(seqs) else ' ' for i in range(len(seqf))]

  return seqf, seqs

def implement_blosum62(seqf, seqs):
  score = 0

  if len(seqf) != len(seqs):
    seqf, seqs = normalization_length(seqf, seqs)

  for i in range(len(seqf)):
    score += blosum62_score(seqf[i], seqs[i])

  return {
      'sequence_1': seqf,
      'sequence_2': seqs,
      'score': score
  }
```
